# 圆的方程

在平面直角坐标系中，以点$O(a,b)$为圆心，以$r$为半径的圆的标准方程是：
$(x-a)^2+(y-b)^2 = r^2$

```python
fig = plt.figure() 
ax = fig.add_subplot(111)
c1 = plt.Circle(xy=(0.0,1.0), radius=1, alpha=0.5)
r1 = plt.Rectangle((0.2,0.75), 0.4,0.15,color="r", alpha=0.3) # 左下起点，长，宽，颜色，alpha透明制

plygon = [[0.0,0.0], [0.35,0.4], [0.2,0.6]]
pgon1 = plt.Polygon(plygon,color="g", alpha=0.5)
ax.add_patch(c1)
ax.add_patch(r1)
ax.add_patch(pgon1)
x,y = 0,0
ax.plot(x,y, "ro")
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

x = ((1,3),(2,5))
y = ((4,7),(6,3))

plt.plot(x[0],y[0],color="r")
plt.plot(x[1],y[1],color="b")

plt.scatter(x[0],y[0],color="b")
plt.scatter(x[1],y[1],color='r')

plt.scatter(x[0],y[0], color="", marker='o', edgecolors="g", s=200)
```
